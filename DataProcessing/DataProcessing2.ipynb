Data Read

```python
path='Data/'
file_list=os.listdir(path)
file_list
```

```python
char=['발생년', '사망자수', '사상자수', '중상자수', '경상자수', '부상신고자수', '발생지시도',
       '발생지시군구', '사고유형_대분류', '경도', '위도']
```

```python
data=[]
for i in file_list :
    print(i);
    d=pd.read_csv(path+i,  encoding = 'utf-8')
    d=d.dropna()
    d=d[char]
    d['부상자수']=d['중상자수']+d['경상자수']+d['부상신고자수']
    d['횟수']=1
    #d.columns=col
    data.append(d)
```

```python
col=['발생년', '사망자수', '사상자수', '중상자수', '경상자수', '부상신고자수', '시도',
       '시군구', '사고유형', '경도', '위도','부상자수','횟수']
```

```python
df=pd.concat(data,ignore_index=True)
df.columns=col
```
