```python
data=[]
for i in file_list :
    print(i);
    d=pd.read_csv(path+i,  encoding = 'utf-8')
    d=d.dropna()
    d=d[char]
    d.columns=col
    d['부상자수']=d['부상자수']+d['경상자수']+d['중상자수']
    d=d.drop('경상자수',1)
    d=d.drop('중상자수',1)
    d=d.replace({'주야':'주'},{'주야':'주간'})
    d=d.replace({'주야':'야'},{'주야':'야간'})
    d=d.replace({'피해자':'0'},{'피해자':'없음'})
    d['횟수']=1
    data.append(d)
```

```python
totalData=pd.concat([data[0],data[1]],axis=0)
for i in range(2,len(data)):
    totalData=pd.concat([totalData,data[i]],axis=0)
hurtData=totalData
hurtData
```

```python
total=len(hurtData)
total
```

```python
test=hurtData.drop('year',axis=1)
test=test.groupby(by=['location'], as_index=False).sum()
test=test.sort_values(by='count', ascending=False)
test=test[['location','count']]
test.set_index(test['location'], inplace=True)
test=test.drop('location',1)
test.columns=['percent']
test=test.T
for i in test.columns :
    test[i]=round(test[i]/total*100,2)
test=test.T
test.to_csv('location.csv',encoding='euc-kr', index=True)
test    
```

```python
col_2=['dayAndNight','count']
dayNight=hurtData[col_2]
dayNight=dayNight.groupby(by=['dayAndNight'], as_index=False).sum()
dayNight.set_index(dayNight['dayAndNight'], inplace=True)
dayNight=dayNight.drop('dayAndNight',1)
dayNight.columns=['percent']
dayNight=dayNight.T
for i in dayNight.columns :
    dayNight[i]=round(dayNight[i]/total*100,2)
dayNight=dayNight.T
dayNight.to_csv('dayAndNight.csv',encoding='euc-kr', index=True)
dayNight
```
