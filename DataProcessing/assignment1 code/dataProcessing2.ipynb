```python
import json
import urllib
import requests
```

```python
def getGPS_KAKAO(address):

    headers = {
        'Content-Type': 'application/json; charset=utf-8',
        'Authorization': 'KakaoAK {}'.format('888b7641561531e72bd804a6c58b0362')
    }
    address = address.encode("utf-8")

    p = urllib.parse.urlencode(
        {
            'query': address
        }
    )

    result = requests.get("https://dapi.kakao.com/v2/local/search/keyword.json", headers=headers, params=p)
    result=result.json()
    longitude = result['documents'][0]['x'] #경도
    latitude = result['documents'][0]['y'] #위도
    return longitude, latitude
```

```python
sta_name=[]
sta_long=[]
sta_lat=[]
for i in am7_TOP30['역명'] :
    sta_name.append(i)
    x,y=getGPS_KAKAO(i)
    sta_long.append(x)
    sta_lat.append(y)
sta_name=pd.DataFrame(sta_name)
sta_long=pd.DataFrame(sta_long)
sta_lat=pd.DataFrame(sta_lat)
sta_am7=pd.concat([sta_name,sta_long,sta_lat],axis=1)
sta_am7.columns=['역명','경도','위도']
sta_am7.to_csv('Data2/sta_am7.csv',encoding='euc-kr',index=False)
sta_am7['time']='AM7'
```
