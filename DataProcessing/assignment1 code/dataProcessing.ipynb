```python
char=['midnight', 'one', 'two','three', 'four', 'five', 'six', 'seven',
       'eight', 'nine', 'ten', 'eleven', 'twelve', 'thirteen',
       'fourteen', 'fifteen',
       'sixteen', 'seventeen', 'eighteen',
       'nineteen', 'twenty', 'twenty_one',
       'twenty_two', 'twenty_three']
```

```python
def addColumns(df) :
    for i in range(0,24) :
        ch=char[i]
        path=ch+'_ride_num'
        path2=ch+'_alight_num'
        
        df=df.astype({
            path : int,
            path2 : int
        })
        
        df[ch]=df[path]+df[path2]
    return df
```

data 읽어오기

```python
data=[]
for i in range(2015,2021) :
    if i == 2020 :
        for j in range(1,9):
            df=pd.read_csv('Data/'+str(i)+'_'+str(j)+'.csv', encoding='CP949')
            df=addColumns(df)
            data.append(df)
    else :
        df=pd.read_csv('Data/'+str(i)+'.csv', encoding='CP949')
        df=addColumns(df)
        data.append(df)
    
data=np.asarray(data)   
```

```python
dataMonthlist=[]
for i in range(2015,2021):
    
    if(i==2020) :
        for j in range(1,9):
            d=df[df['use_mon'].isin([str(i)+str(0)+str(j)])]
            dataMonthlist.append(d)
            
    else :
        for j in range(1,10):
            d=df[df['use_mon'].isin([str(i)+str(0)+str(j)])]
            dataMonthlist.append(d)
        for j in range(10,13):
            d=df[df['use_mon'].isin([str(i)+str(j)])]
            dataMonthlist.append(d)
            
dataMonthlist=np.asarray(dataMonthlist)
```

```python
import json
import urllib
import requests
```

```python
def getGPS_KAKAO(address):

    headers = {
        'Content-Type': 'application/json; charset=utf-8',
        'Authorization': 'KakaoAK {}'.format('888b7641561531e72bd804a6c58b0362')
    }
    address = address.encode("utf-8")

    p = urllib.parse.urlencode(
        {
            'query': address
        }
    )

    result = requests.get("https://dapi.kakao.com/v2/local/search/keyword.json", headers=headers, params=p)
    result=result.json()
    longitude = result['documents'][0]['x'] #경도
    latitude = result['documents'][0]['y'] #위도
    return longitude, latitude
```
